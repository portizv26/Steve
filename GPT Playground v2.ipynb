```python
import random
from tqdm import tqdm

# ids = random.sample(dfi[dfi.name_component == 'MOTOR DIESEL'].iloc[-5000:].index.tolist(), k=10)
ids = [330624]
l = [generate_comment(id) for id in tqdm(ids, desc='Generating comments', unit='comment')]
```

```python
dfh = pd.DataFrame()
for i in range(len(ids)):

    t3 = df2[df2.id_sample == ids[i]]
    dfk = pd.merge(t3,
                   pd.DataFrame({'id_sample':ids[i],
                                 'comment T=0.2':[l[i][0][0]],
                                 'comment T=0.5':[l[i][0][1]],
                                 'comment T=0.8':[l[i][0][2]]}), 
                    on = 'id_sample')
    dfh = pd.concat([dfh, dfk])
dfh.to_excel('merged_data.xlsx')
```
